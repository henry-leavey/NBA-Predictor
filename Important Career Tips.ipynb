## Career Decision-Making Use Cases

| Scenario | Best Model Family | Why |
| :- | :- | :- |
| Predicting continuous numeric output (sales, price, points) | Linear Regression / Gradient Boosting | Handles regression well |
| Binary classification (fraud / churn / disease) | Logistic Regression / Random Forest / XGBoost | Balance interpretability and accuracy |
| Image classification | CNN | Extracts spatial patterns |
| Text classification | Transformer / Naive Bayes / Logistic Regression | Handles language features |
| Customer segmentation | K-Means / DBSCAN | Groups similar data |
| Reducing features / visualizing | PCA / t-SNE / UMAP | Dimensionality reduction |
| Forecasting (time-series) | RNN / LSTM / Prophet | Handles sequential dependencies |
| Tabular data (structured business data) | Random Forest / XGBoost / LightGBM | Reliable performance on real-world data |
